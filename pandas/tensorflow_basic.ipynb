```python
import tensorflow as tf

# 1. Tạo tensor x kích thước (3, 2) trong khoảng [5, 10]
x = tf.random.uniform(shape=[3, 2], minval=5, maxval=10, dtype=tf.float32)

# 2. Tạo tensor y kích thước (2, 3) trong khoảng [15, 20]
y = tf.random.uniform(shape=[2, 3], minval=15, maxval=20, dtype=tf.float32)

# 3. Lấy các giá trị đường chéo của x
# tf.linalg.diag_part sẽ trả về các phần tử x[0,0] và x[1,1]
diag_x = tf.linalg.diag_part(x)

# 4. Thực hiện phép cộng x + y
# Vì x(3,2) và y(2,3) lệch kích thước, ta cần chuyển vị y (transpose) để thành (3,2)
y_transposed = tf.transpose(y)
sum_result = x + y_transposed

print("Tensor x:\n", x.numpy())
print("\nĐường chéo của x:\n", diag_x.numpy())
print("\nTensor y (sau khi chuyển vị để cộng):\n", y_transposed.numpy())
print("\nKết quả x + y_transpose:\n", sum_result.numpy())
```

```python
import tensorflow as tf

# 1. Hàm tạo 2 tensor ngẫu nhiên kích thước (3, 2, 2)
def create_tensors():
    # Sử dụng phân phối chuẩn (Normal) cho các giá trị ngẫu nhiên
    x = tf.random.normal(shape=(3, 2, 2))
    y = tf.random.normal(shape=(3, 2, 2))
    return x, y

# 2. Định nghĩa các hàm toán học và chuyển vào Graph bằng @tf.function
@tf.function
def compute_ops(x, y):
    add_res = tf.add(x, y)
    sub_res = tf.subtract(x, y)
    mul_res = tf.multiply(x, y) # Nhân từng phần tử (Element-wise)
    div_res = tf.divide(x, y)   # Chia từng phần tử

    return add_res, sub_res, mul_res, div_res

# --- Thực thi trong Colab ---

# Bước 1: Tạo dữ liệu
tensor_x, tensor_y = create_tensors()

# Bước 2: Tính toán (Lần gọi đầu tiên sẽ "trace" hàm vào Graph)
addition, subtraction, multiplication, division = compute_ops(tensor_x, tensor_y)

# In kết quả kiểm tra
print("Tensor X:\n", tensor_x.numpy())
print("\nKết quả phép cộng (X + Y):\n", addition.numpy())
```

```python
import tensorflow as tf

def build_sum_model(input_shape):
    # Định nghĩa 2 đầu vào riêng biệt cho x và y
    input_x = tf.keras.Input(shape=input_shape, name="input_x")
    input_y = tf.keras.Input(shape=input_shape, name="input_y")

    # Sử dụng lớp Lambda hoặc phép cộng trực tiếp để tính tổng
    # Trong TF, bạn có thể dùng layer Add để tối ưu cho model
    sum_layer = tf.keras.layers.Add()([input_x, input_y])

    # Khởi tạo model với 2 đầu vào và 1 đầu ra
    model = tf.keras.Model(inputs=[input_x, input_y], outputs=sum_layer)

    return model

# Giả sử tensor có kích thước (2, 2)
input_shape = (2, 2)
sum_model = build_sum_model(input_shape)

# Hiển thị cấu trúc model
sum_model.summary()
```

```python
import tensorflow as tf

def build_math_model(input_shape):
    # 1. Định nghĩa các cổng đầu vào (Input layers)
    input_x = tf.keras.Input(shape=input_shape, name="input_x")
    input_y = tf.keras.Input(shape=input_shape, name="input_y")

    # 2. Thực hiện phép trừ (Hiệu)
    # Ta dùng layer Subtract
    diff = tf.keras.layers.Subtract(name="subtraction")([input_x, input_y])

    # 3. Thực hiện phép nhân (Tích - Element-wise)
    # Ta dùng layer Multiply
    prod = tf.keras.layers.Multiply(name="multiplication")([input_x, input_y])

    # 4. Khởi tạo model với 2 đầu vào và một danh sách các đầu ra
    model = tf.keras.Model(inputs=[input_x, input_y], outputs=[diff, prod])

    return model

# Giả sử kích thước tensor là (3, 2)
input_shape = (3, 2)
math_model = build_math_model(input_shape)

# Hiển thị cấu trúc mô hình
math_model.summary()
```

```python
import tensorflow as tf

def build_total_area_model():
    # 1. Khai báo các đầu vào (Input layers)
    a = tf.keras.Input(shape=(1,), name="length")  # Chiều dài
    b = tf.keras.Input(shape=(1,), name="width")   # Chiều rộng
    c = tf.keras.Input(shape=(1,), name="height")  # Chiều cao

    # 2. Tính Diện tích xung quanh: S_xq = (a + b) * c * 2
    sum_ab = tf.keras.layers.Add(name="a_plus_b")([a, b])
    # Dùng Lambda để nhân với c và nhân với 2
    s_xq = tf.keras.layers.Lambda(lambda x: x[0] * x[1] * 2, name="S_xung_quanh")([sum_ab, c])

    # 3. Tính Diện tích 2 mặt đáy: S_day = (a * b) * 2
    s_day = tf.keras.layers.Lambda(lambda x: x[0] * x[1] * 2, name="S_2_day")([a, b])

    # 4. Tính Diện tích toàn phần: S_tp = S_day + S_xq
    s_tp = tf.keras.layers.Add(name="S_toan_phan")([s_day, s_xq])

    # 5. Khởi tạo Model
    model = tf.keras.Model(inputs=[a, b, c], outputs=s_tp)
    return model

# Khởi tạo mô hình
area_model = build_total_area_model()

# --- CHẠY THỬ NGHIỆM ---
# Giả sử a=5, b=3, c=4
# S_xq = (5+3)*4*2 = 64
# S_tp = (5*3)*2 + 64 = 30 + 64 = 94
test_a = tf.constant([[5.0]])
test_b = tf.constant([[3.0]])
test_c = tf.constant([[4.0]])

result = area_model.predict({'length': test_a, 'width': test_b, 'height': test_c})

print(f"--- KẾT QUẢ ---")
print(f"Chiều dài: 5, Chiều rộng: 3, Chiều cao: 4")
print(f"Diện tích toàn phần tính từ Model: {result[0][0]}")

# Hiển thị cấu trúc mô hình
area_model.summary()
```

```python
import tensorflow as tf

def build_trapezoid_model():
    # 1. Khai báo 4 đầu vào là độ dài các cạnh
    ad = tf.keras.Input(shape=(1,), name="AD")
    bc = tf.keras.Input(shape=(1,), name="BC")
    ab = tf.keras.Input(shape=(1,), name="AB")
    cd = tf.keras.Input(shape=(1,), name="CD")

    # 2. Kiểm tra điều kiện 1: AD == BC
    # Cần ép kiểu hoặc chỉ định dtype cho Lambda khi làm việc với logic
    cond1 = tf.keras.layers.Lambda(lambda x: tf.equal(x[0], x[1]),
                                   name="check_AD_BC", dtype='bool')([ad, bc])

    # 3. Kiểm tra điều kiện 2: AB == CD
    cond2 = tf.keras.layers.Lambda(lambda x: tf.equal(x[0], x[1]),
                                   name="check_AB_CD", dtype='bool')([ab, cd])

    # 4. Kết hợp cả hai điều kiện (Logical AND)
    is_parallel = tf.keras.layers.Lambda(lambda x: tf.logical_and(x[0], x[1]),
                                         name="AD_parallel_BC", dtype='bool')([cond1, cond2])

    # 5. Khởi tạo Model
    model = tf.keras.Model(inputs=[ad, bc, ab, cd], outputs=is_parallel)
    return model

# Khởi tạo mô hình
model = build_trapezoid_model()

# --- KIỂM TRA THỬ NGHIỆM ---
test_data = {
    "AD": tf.constant([[5.0]]),
    "BC": tf.constant([[5.0]]),
    "AB": tf.constant([[3.0]]),
    "CD": tf.constant([[3.0]])
}

result = model.predict(test_data)
print(f"Kết quả kiểm tra (Song song?): {result[0][0]}")
```

```python
import tensorflow as tf

def build_base_equal_model():
    # 1. Định nghĩa các đầu vào
    # ad, bc: Độ dài hai cạnh bên
    # is_parallel: Một tensor kiểu Boolean (True/False) đại diện cho điều kiện AD // BC
    ad = tf.keras.Input(shape=(1,), name="AD_length")
    bc = tf.keras.Input(shape=(1,), name="BC_length")
    is_parallel = tf.keras.Input(shape=(1,), dtype=tf.bool, name="AD_parallel_BC")

    # 2. Kiểm tra điều kiện độ dài: AD == BC
    # Thêm dtype='bool' để Keras xử lý được kết quả logic
    equal_length = tf.keras.layers.Lambda(
        lambda x: tf.equal(x[0], x[1]), name="check_length", dtype='bool'
    )([ad, bc])

    # 3. Kết hợp hai điều kiện bằng phép AND logic
    # AB = CD <=> (AD == BC) AND (AD // BC)
    proof = tf.keras.layers.Lambda(
        lambda x: tf.logical_and(x[0], x[1]), name="proof_AB_equal_CD", dtype='bool'
    )([equal_length, is_parallel])

    # 4. Khởi tạo Model
    model = tf.keras.Model(inputs=[ad, bc, is_parallel], outputs=proof)
    return model

# Khởi tạo mô hình
model_proof = build_base_equal_model()

# --- KIỂM TRA THỬ NGHIỆM ---
# Giả sử AD = 5, BC = 5 và đã biết AD // BC (True)
test_input = {
    "AD_length": tf.constant([[5.0]]),
    "BC_length": tf.constant([[5.0]]),
    "AD_parallel_BC": tf.constant([[True]])
}

result = model_proof.predict(test_input)
print(f"Kết quả chứng minh (AB = CD?): {result[0][0]}")
```

```python
import tensorflow as tf

def build_midpoint_proof_model():
    # 1. Định nghĩa các đầu vào theo mô tả
    am = tf.keras.Input(shape=(1,), name="AM")
    mb = tf.keras.Input(shape=(1,), name="MB")
    mn_parallel_bc = tf.keras.Input(shape=(1,), dtype=tf.bool, name="MN_parallel_BC")

    # 2. Kiểm tra điều kiện 1: M có là trung điểm của AB không? (AM == MB)
    # Thêm dtype='bool' để Keras xử lý được kết quả logic
    is_m_midpoint = tf.keras.layers.Lambda(
        lambda x: tf.equal(x[0], x[1]), name="check_M_midpoint", dtype='bool'
    )([am, mb])

    # 3. Kết hợp điều kiện M là trung điểm VÀ MN // BC
    # Theo định lý: Nếu một đường thẳng đi qua trung điểm một cạnh của tam giác
    # và song song với cạnh thứ hai thì nó đi qua trung điểm cạnh thứ ba.
    proof_result = tf.keras.layers.Lambda(
        lambda x: tf.logical_and(x[0], x[1]), name="proof_NA_equal_NC", dtype='bool'
    )([is_m_midpoint, mn_parallel_bc])

    # 4. Khởi tạo Model
    model = tf.keras.Model(inputs=[am, mb, mn_parallel_bc], outputs=proof_result)
    return model

# Khởi tạo mô hình
model_triangle = build_midpoint_proof_model()

# --- CHẠY THỬ NGHIỆM TRÊN COLAB ---
# Giả sử AM = 4, MB = 4 (M là trung điểm) và MN // BC là đúng (True)
test_inputs = {
    "AM": tf.constant([[4.0]]),
    "MB": tf.constant([[4.0]]),
    "MN_parallel_BC": tf.constant([[True]])
}

result = model_triangle.predict(test_inputs)

print(f"--- KẾT QUẢ CHỨNG MINH ---")
print(f"Kết luận NA = NC là: {result[0][0]}")
```
