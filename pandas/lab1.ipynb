```python
import numpy as np

# 1. Tạo mảng 20 phần tử ngẫu nhiên trong khoảng [0, 20]
# Chúng ta dùng randint để tạo số nguyên
arr = np.random.randint(0, 21, 20)

print("Mảng ban đầu:")
print(arr)

# 2. Đổi dấu những số nằm trong khoảng [9, 15]
# Điều kiện: (phần tử >= 9) VÀ (phần tử <= 15)
# Trong Numpy, toán tử VÀ cho mảng là dấu '&'
mask = (arr >= 9) & (arr <= 15)

# Áp dụng mặt nạ (mask) để nhân các phần tử thỏa mãn với -1
arr[mask] = arr[mask] * -1

print("\nMảng sau khi đổi dấu các số trong khoảng [9, 15]:")
print(arr)
```

```python
import numpy as np

# 1. Tạo ma trận 10x10 toàn số 1
# Chúng ta dùng dtype=int để ma trận nhìn gọn gàng hơn
matrix = np.ones((10, 10), dtype=int)

# 2. Thay đổi các phần tử bên trong thành số 0
# Sử dụng slicing [hàng_bắt_đầu : hàng_kết_thúc, cột_bắt_đầu : cột_kết_thúc]
# Chúng ta bỏ hàng đầu (index 0) và hàng cuối (index -1) -> dùng 1:-1
# Chúng ta bỏ cột đầu (index 0) và cột cuối (index -1) -> dùng 1:-1
matrix[1:-1, 1:-1] = 0

print("Ma trận 10x10 có biên bằng 1, bên trong bằng 0:")
print(matrix)
```

```python
import numpy as np

# 1. Tạo danh sách các giá trị cho đường chéo chính
diagonal_values = [1, 2, 3, 4, 5]

# 2. Tạo ma trận 5x5 từ danh sách trên
# Hàm diag sẽ tự động tạo ma trận vuông có kích thước tương ứng với độ dài mảng
matrix = np.diag(diagonal_values)

print("Ma trận 5x5 với đường chéo chính là 1, 2, 3, 4, 5:")
print(matrix)

# Kiểm tra kích thước ma trận
print(f"\nKích thước ma trận: {matrix.shape}")
```

```python
import pandas as pd
import numpy as np

# 1. Tạo DataFrame 4x3 với các số ngẫu nhiên từ 1 đến 100
# Chúng ta đặt tên cột là 'A', 'B', 'C' để dễ thao tác
df = pd.DataFrame(np.random.randint(1, 101, size=(4, 3)),
                  columns=['Cột A', 'Cột B', 'Cột C'])

print("DataFrame ban đầu:")
print(df)

# 2. Sắp xếp DataFrame theo 'Cột A'
# ascending=True: Sắp xếp tăng dần (mặc định)
# ascending=False: Sắp xếp giảm dần
df_sorted = df.sort_values(by='Cột A', ascending=True)

print("\nDataFrame sau khi sắp xếp tăng dần theo 'Cột A':")
print(df_sorted)

# 3. Sắp xếp theo nhiều cột (Ví dụ: B tăng dần, nếu bằng nhau thì C giảm dần)
df_multi_sorted = df.sort_values(by=['Cột B', 'Cột C'], ascending=[True, False])

print("\nSắp xếp theo Cột B (tăng) và Cột C (giảm):")
print(df_multi_sorted)
```

```python
import pandas as pd

# 1. Khởi tạo 2 Series theo yêu cầu
sr1 = pd.Series([1, 2, 3, 4, 5])
sr2 = pd.Series([2, 4, 6, 8, 10])

print("Series 1 (sr1):")
print(sr1)
print("\nSeries 2 (sr2):")
print(sr2)

# 2. Tìm các phần tử trong sr1 mà không có trong sr2
# Bước 1: sr1.isin(sr2) trả về True nếu phần tử của sr1 nằm trong sr2
# Bước 2: Dấu ~ dùng để đảo ngược kết quả (True thành False và ngược lại)
result = sr1[~sr1.isin(sr2)]

print("\nItems of sr1 not present in sr2:")
print(result)
```

```python
import numpy as np
import pandas as pd

# 1. Tạo 2 Series ngẫu nhiên có 20 phần tử trong khoảng [1, 15]
# Dùng randint để tạo số nguyên
sr1 = pd.Series(np.random.randint(1, 16, 20))
sr2 = pd.Series(np.random.randint(1, 16, 20))

print("Series 1 (đã loại trùng để dễ nhìn):", np.unique(sr1.values))
print("Series 2 (đã loại trùng để dễ nhìn):", np.unique(sr2.values))
print("-" * 40)

# 2. Tìm phần GIAO (Intersect) - Những số có mặt ở cả hai
giao = np.intersect1d(sr1, sr2)

# 3. Tìm phần HỘI (Union) - Tất cả các số của cả hai, không trùng lặp
hoi = np.union1d(sr1, sr2)

print("GIAO (Các phần tử xuất hiện ở cả hai Series):")
print(giao)

print("\nHỘI (Tất cả các phần tử duy nhất của cả hai Series):")
print(hoi)
```

```python
import pandas as pd

# 1. Khởi tạo 2 Series theo yêu cầu
# Series x chứa các số từ 0 đến 9
x = pd.Series(range(10), name='x')

# Series y chứa các ký tự từ p đến y
y = pd.Series(['p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y'], name='y')

print("Series x gốc:")
print(x.head(3), "...", sep="\n") # In 3 dòng đầu để minh họa

print("\nSeries y gốc:")
print(y.head(3), "...", sep="\n")
print("-" * 30)

# 2. Thêm theo chiều dọc (Vertical)
# axis=0 (mặc định): Nối đuôi nhau tạo thành một Series dài hơn (20 phần tử)
vertical_concat = pd.concat([x, y], axis=0)

# 3. Thêm theo chiều ngang (Horizontal)
# axis=1: Kết hợp lại tạo thành một DataFrame (10 hàng, 2 cột)
horizontal_concat = pd.concat([x, y], axis=1)

print("Kết quả theo chiều dọc (Vertical - tạo Series dài):")
print(vertical_concat)

print("\nKết quả theo chiều ngang (Horizontal - tạo DataFrame):")
print(horizontal_concat)
```
